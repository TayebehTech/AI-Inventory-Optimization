```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
```

```python
# Sample data
data = {
    "ds": pd.date_range(start="2023-01-01", periods=20, freq="W"),
    "y": [100, 120, 130, 125, 140, 135, 145, 160, 155, 170, 180, 175, 185, 190, 200, 210, 220, 230, 240, 250]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Prepare data for Prophet
df_prophet = df.reset_index(drop=True)  # Ensure index is reset
df_prophet.columns = ["ds", "y"]  # Prophet requires specific column names
```

```python
# Train Prophet model
prophet_model = Prophet()
prophet_model.fit(df_prophet)
```

```python
# Make future predictions
future = prophet_model.make_future_dataframe(periods=5, freq="W")
forecast = prophet_model.predict(future)
```

```python
# Plot results
prophet_model.plot(forecast)
plt.title("Prophet Demand Forecasting")
plt.show()
```
